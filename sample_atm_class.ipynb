We will try to make a sample class for ATM machine. We will include following menu in the class, so that user can use it according to their need.
1.   Enter 1 to create pin
2.   Enter 2 to change your pin
3.   Enter 3 to check your balance
4.   Enter 4 to withdraw balance
5.   Enter 5 to deposite balanced
6.   Enter 6 to exit

```python
class Atm:
  def __init(self):
    self.account_name = ' '
    self.account_number = ' '
    self.account_balance = ' '
    self.atm_pin = ' '
    self.menu()
  
  #A function to add details in the account
  def add_details(self):
    '''
    This function enables to add details of the account
    '''
    account_name = input('Enter account holder name: ')
    self.account_name = account_name
    account_number = int(input('Enter your account number: '))
    self.account_number = account_number
    account_balance = int(input('Enter account balance: '))
    self.account_balance = account_balance
    print('Account created successfully')

  #A function to see all the menu
  def menu(self):
    '''
    This function will give all the menu
    '''
    user_input = input('''
    1. Enter 1 to create pin
    2. Enter 2 to change your pin
    3. Enter 3 to check your balance
    4. Enter 4 to withdraw balance
    5. Enter 5 to deposite balanced
    6. Enter 6 to check full details
    7. Enter 7 to exit
    ''')
    if user_input == '1':
      self.create_pin()
    elif user_input == '2':
      self.change_pin()
    elif user_input == '3':
      self.check_balance()
    elif user_input == '4':
      self.withdraw_balance()
    elif user_input == '5':
      self.deposite_balance()
    elif user_input == '6':
      self.check_details()
    else:
      exit()
    
    #A function to create pin
  def create_pin(self):
    '''
      It will allows to create new pin
    '''
    atm_pin = int(input('Enter pin of your choice: '))
    self.atm_pin = atm_pin
    print('Atm pin created successfully')
    self.menu()

  #A function to change atm pin
  def change_pin(self):
    '''
      It will allows to change pin
    '''
    old_pin = int(input('Enter old pin: '))
    if old_pin == self.atm_pin:
      new_pin = int(input('Enter new pin: '))
      self.atm_pin = new_pin
      print('Congratulation, your pin has been changed successfully')
      self.menu()
    else:
      print('You have entered wrong atm pin')
      self.menu()

    #A function to check balance
  def check_balance(self):
    '''
      It will allow user to check their account balance
    '''
    atm_pin = int(input('Enter your atm pin: '))
    if atm_pin == self.atm_pin:
      print(f'Your account balance is: {self.account_balance}')
      self.menu()
    else:
      print('You have entered wrong atm pin')
      self.menu()

  #A function to withdraw balance
  def withdraw_balance(self):
    '''
    It will allow user to withdraw balance from their account
    '''
    atm_pin = int(input('Enter your atm pin: '))
    if atm_pin == self.atm_pin:
      enter_amount = int(input('Enter amount to withdraw: '))
      if enter_amount <= self.account_balance:
        self.account_balance = self.account_balance - enter_amount
        print(f'Withdraw from account is successful \n
        Your current balance is: {self.account_balance}')
        self.menu()
      else:
        print('Entered balance is lower than your account balance')
        self.menu()
    else:
      print('You have entered wrong atm pin')
      self.menu()

  #A function to deposite balance
  def deposite_balance(self):
    '''
      It will allow user to deposite balance in their account
    '''
    atm_pin = int(input('Enter your atm pin: '))
    if atm_pin == self.atm_pin:
      enter_amount = int(input('Enter amount to withdraw: '))
      self.account_balance = self.account_balance + enter_amount
      print(f'Deposite in your account is successful \n
      Your current balance is: {self.account_balance}')
      self.menu()
    else:
      print('You have entered wrong atm pin')
      self.menu()

  #A function to check full details of the account
  def check_details(self):
    '''
    This function enables to check full details of the account
    '''
    enter_pin = int(input('Enter your pin: '))
    if enter_pin == self.atm_pin:
      print(f'Account name: {self.account_name}')
      print(f'Account number: {self.account_number}')
      print(f'Current balance: {self.account_balance}')
      self.menu()
    else:
      print('Entered wrong atm pin')
      self.menu()
```
